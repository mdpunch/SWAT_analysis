```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
import seaborn as sns

#sns.set_context("paper", rc={"font.size":22,"axes.titlesize":18,"axes.labelsize":26})   
sns.set(font_scale=1.5)
sns.set_style("darkgrid",
              {'grid.linestyle': '--',  'grid.color': 'lightgrey'})
plt.rcParams.update({'font.size': 16})
```

```python
sim_rate = np.array([round(np.power(2,np.log2(1.46)*i)*128) for i in np.arange(8)])*1000
print(sim_rate/1000)
```

```python
dead = 400e-9
```

```python
event_rate = sim_rate*np.exp(-dead*sim_rate)
print(event_rate/1000)
```

```python
# TATS_SWAT_analysis/tmp_highfreq.txt
noted_rate = np.array([118, 165, 225, 306, 411, 541, 600, 880])*1000
print(noted_rate/1000)
```

```python
#plt.plot(noted_rate,sim_rate)
plt.plot(noted_rate,event_rate)
plt.plot([120000,900000],[120000,900000])
```

```python
plt.plot(noted_rate,event_rate/noted_rate)
```

```python
multi_of_8 = [7.9, 7.7, 7.4, 7.1, 6.9, 6.8, 6.7, 6.6]
```

```python
plt.plot(noted_rate/1000, multi_of_8,"o",label="High range")
plt.plot(64000/1000,8,"o",label="Low range")
plt.legend(loc="lower right")
plt.xlabel("Event rate (kHz)")
plt.ylabel("Average SWAT Multiplicity (of 8)")
plt.xlim(xmin=0)
plt.ylim(ymin=2)
None
```
